```python
df = pd.read_csv("/Users/kseniadudiy/Downloads/texts_for_proj.csv", sep=",")  
df
```

```python
sentss = []
for item in df['text']:
    new_item = []
    doc = nlp(item)
    assert doc.has_annotation("SENT_START")
    for sent in doc.sents:
        new_item.append(sent.text)
    sentss.append(new_item)
df['sentences'] = sentss
df
```

```python
s_pos = []
for item in df['sentences']:
    item_pos = []
    for element in item:
        sent_pos = []
        els = nlp(element)
        for token in els:
            if token.pos_ != 'SPACE' and token.pos_ != 'PUNCT' and token.text.lower() != '-' and token.text.lower() != '!' and token.text.lower() != '?' and token.text.lower() != '.' and token.text.lower() != ',':
                sent_pos.append([token.text.lower(), token.lemma_, token.pos_])
        item_pos.append(sent_pos)
    s_pos.append(item_pos)
df['pos_tagged'] = s_pos
df
```

```python
import re
def exact_form(search_item):
    forms = []
    search_item = search_item.replace("\"", '')
    for index, row in df.iterrows():
        for item in row['pos_tagged']:
            a = row['pos_tagged'].index(item)
            for element in item:
                if element[0] == search_item:
                    theme = row['topic']
                    raw_sentence = row['sentences'][a]
                    forms.append([theme, raw_sentence, element[0]])
    return(forms)
def lemma(search_item):
    forms = []
    for index, row in df.iterrows():
        for item in row['pos_tagged']:
            a = row['pos_tagged'].index(item)
            for element in item:
                if element[1] == search_item:
                    theme = row['topic']
                    raw_sentence = row['sentences'][a]
                    forms.append([theme, raw_sentence, element[0]])
    return(forms)
def pos(search_item):
    forms = []
    for index, row in df.iterrows():
        for item in row['pos_tagged']:
            a = row['pos_tagged'].index(item)
            for element in item:
                if element[2] == search_item:
                    theme = row['topic']
                    raw_sentence = row['sentences'][a]
                    forms.append([theme, raw_sentence, element[0]])
    return(forms[:50])
def word_pos(search_item):
    forms = []
    n_sr = re.split('\+', search_item)
    for index, row in df.iterrows():
        for item in row['pos_tagged']:
            a = row['pos_tagged'].index(item)
            for element in item:
                if element[1] == n_sr[0] and element[2] == n_sr[1]:
                    theme = row['topic']
                    raw_sentence = row['sentences'][a]
                    forms.append([theme, raw_sentence, element[0]])
    return(forms)
```

```python
def sentence_checker(sentence, search_q, previous_q):
    sent_tokens = []
    els = nlp(sentence)
    for token in els:
        if token.pos_ != 'SPACE' and token.pos_ != 'PUNCT' and token.text.lower() != '-' and token.text.lower() != '!' and token.text.lower() != '?' and token.text.lower() != '.' and token.text.lower() != ',':
            sent_tokens.append([token.text.lower(), token.lemma_, token.pos_])
    if re.search('\+', search_q) != None:
        n_sr = re.split('\+', search_q)
        for item in sent_tokens:
            ind = sent_tokens.index(item)
            new_it = sent_tokens[ind-1]
            if ind >= 1:
                if item[1] == n_sr[0] and (item[2] == n_sr[1] and re.search(previous_q, new_it[0]) != None or re.search(previous_q, new_it[1]) != None or re.search(previous_q, new_it[2]) != None):
                    return([sentence, item[0]])
    elif re.search('\"', search_q) != None:
        search_q = search_q.replace("\"", '')
        for item in sent_tokens:
            ind = sent_tokens.index(item)
            new_it = sent_tokens[ind-1]
            if ind >= 1:
                if item[0] == search_q and (re.search(previous_q, new_it[0]) != None or re.search(previous_q, new_it[1]) != None or re.search(previous_q, new_it[2]) != None):
                    return([sentence, item[0]])
    elif re.search('[A-Z]+', search_q) != None and re.search('\+', search_q) == None:
        for item in sent_tokens:
            ind = sent_tokens.index(item)
            new_it = sent_tokens[ind-1]
            if ind >= 1:
                if item[2] == search_q and (re.search(previous_q, new_it[0]) != None or re.search(previous_q, new_it[1]) != None or re.search(previous_q, new_it[2]) != None):
                    return([sentence, item[0]])
    else:
        for item in sent_tokens:
            ind = sent_tokens.index(item)
            new_it = sent_tokens[ind-1]
            if ind >= 1:
                if item[1] == search_q and (re.search(previous_q, new_it[0]) != None or re.search(previous_q, new_it[1]) != None or re.search(previous_q, new_it[2]) != None):
                    return([sentence, item[0]])
                
```

```python
def search_parcer(search_element):
    search_element = input()
    split_search = re.split(' ', search_element)
    if len(split_search) > 3:
        return('извините, ваш запрос слишком длинный! попробуйте запрос из трех слов или меньше')
    elif len(split_search) == 0:
        return('вы ничего не написали :(')
    elif len(split_search) == 1:
        n = split_search[0]
        if re.search('\+', n) != None:
            return(word_pos(n))
        elif re.search('\"', n) != None:
            return(exact_form(n))
        elif re.search('[A-Z]+', n) != None and re.search('\+', n) == None:
            return(pos(n))
        else:
            return(lemma(n))
    elif len(split_search) == 2:
        m = split_search[0]
        q = split_search[1]
        if re.search('\+', m) != None:
            wpos_2 = []
            for item in word_pos(m):
                if sentence_checker(item[1], q, item[2]) is not None:
                    wpos_2.append([item[0], sentence_checker(item[1], q, item[2])])
            return(wpos_2)
        elif re.search('\"', m) != None:
            form_2 = []
            for item in exact_form(m):
                if sentence_checker(item[1], q, item[2]) is not None:
                    form_2.append([item[0], sentence_checker(item[1], q, item[2])])
            return(form_2)
        elif re.search('[A-Z]+', m) != None and re.search('\+', m) == None:
            pos_2 = []
            for item in pos(m):
                if sentence_checker(item[1], q, item[2]) is not None:
                    pos_2.append([item[0], sentence_checker(item[1], q, item[2])])
            return(pos_2)
        else:
            lem_2 = []
            for item in lemma(m):
                if sentence_checker(item[1], q, item[2]) is not None:
                    lem_2.append([item[0], sentence_checker(item[1], q, item[2])])
            return(lem_2)
    elif len(split_search) == 3:
        r = split_search[0]
        s = split_search[1]
        t = split_search[2]
        if re.search('\+', r) != None:
            list_1 = []
            wpos_3 = []
            for item in word_pos(r):
                if sentence_checker(item[1], s, item[2]) is not None:
                    list_1.append(sentence_checker(item[1], s, item[2]))
                    for element in list_1:
                        if sentence_checker(element[0], t, element[1]) is not None:
                            wpos_3.append([item[0], sentence_checker(element[0], t, element[1])])
            return(wpos_3)
        elif re.search('\"', r) != None:
            list_2 = []
            form_3 = []
            for item in exact_form(r):
                if sentence_checker(item[1], s, item[2]) is not None:
                    list_2.append(sentence_checker(item[1], s, item[2]))
                    for element in list_2:
                        if sentence_checker(element[0], t, element[1]) is not None:
                            form_3.append([item[0], sentence_checker(element[0], t, element[1])])
            return(form_3)
        elif re.search('[A-Z]+', r) != None and re.search('\+', m) == None:
            list_3 = []
            pos_3 = []
            for item in pos(r):
                if sentence_checker(item[1], s, item[2]) is not None:
                    list_3.append(sentence_checker(item[1], s, item[2]))
                    for element in list_3:
                        if sentence_checker(element[0], t, element[1]) is not None:
                            pos_3.append([item[0], sentence_checker(element[0], t, element[1])])
            return(pos_3)
        else:
            list_4 = []
            lem_3 = []
            for item in lemma(r):
                if sentence_checker(item[1], s, item[2]) is not None:
                    list_4.append(sentence_checker(item[1], s, item[2]))
                    for element in list_4:
                        if sentence_checker(element[0], t, element[1]) is not None:
                            lem_3.append([item[0], sentence_checker(element[0], t, element[1])])
            return(lem_3)
```

```python
def prettifier(data):
    text_names = []
    sentences = []
    for element in data:
        if len(element) == 3:
            text_names.append(element[0])
            sentences.append(element[1])
        elif len(element) == 2:
            text_names.append(element[0])
            sentences.append(element[1][0])
    dictt = {"название текста":text_names,
            "предложение":sentences}
    df = pd.DataFrame(dictt)
    return(df)
```

```python
def main(data):
    data = search_parcer(search_element)
    return(prettifier(data))
```

Поиск находится в ячейке сразу под этой. При нажатии "запуск" вам будет предложено окошко, в которое вы можете ввести свой запрос. Запрос должен строиться по следующим правилам:
Базовые правила.
- Все слова и леммы пишутся, начиная со строчной буквы
- Нет возможности искать пунктуационные элементы
- Запрос должен состоять из 1-3 элементов.
Оформление поиска:
- Лемма: единица без кавычек (например, день)
- Специфическая словоформа: единица в кавычках (“день”)
- Лемма с спецификацией части речи: единица+тег (день+NOUN)
- Часть речи: тег(NOUN)
Несколько единиц поиска должны быть разделены единичным пробелом.
Примеры:
- добрый день
- “добрый” день
- ADJ “день” NOUN
